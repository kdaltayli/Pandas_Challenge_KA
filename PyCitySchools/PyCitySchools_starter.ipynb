```python
# Dependencies and Setup
import pandas as pd
import numpy as np
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Create an overview table that summarizes key metrics about each school, including:
# School Name
school_summary=school_data_complete.groupby(["school_name","type"]).mean()

school_summary
```

```python
# Total Student
# school_summary["Total Student"]=school_data_complete["school_name"].value_counts().values
# school_summary
school_summary["Total Student"]=school_data_complete.groupby(["school_name","type"]).count()["Student ID"]
school_summary.head()
```

```python
# Total Budget
school_summary["Total School Budget"]=school_summary["budget"]
school_summary["Total School Budget"]=school_summary["Total School Budget"].map("${:,.2f}".format)
# Per Student Budget
school_summary["Per Student Budget"]=school_summary["budget"]/school_summary["size"]
school_summary["Per Student Budget"]=school_summary["Per Student Budget"].map("${:,.2f}".format)
school_summary
```

```python
# Average Math Score
school_summary["Average Math Score"]=school_summary["math_score"]
del school_summary["math_score"]
# Average Reading Score
school_summary["Average Reading Score"]=school_summary["reading_score"]
del school_summary["reading_score"]
school_summary
```

```python
# %Passing Math
pass_math=school_data_complete.loc[school_data_complete["math_score"]>=70]

school_summary["Pass Math student"]=pass_math.groupby(["school_name","type"]).count()["Student ID"]
school_summary["Pass Math student"]=round((school_summary["Pass Math student"]/school_summary["Total Student"])*100,4)
school_summary
```

```python
# %Passing Reading
pass_reading=school_data_complete.loc[school_data_complete["reading_score"]>=70]

school_summary["Pass Reading student"]=pass_reading.groupby(["school_name","type"]).count()["Student ID"]
school_summary["Pass Reading student"]=round((school_summary["Pass Reading student"]/school_summary["Total Student"])*100,4)
school_summary
```

```python
# % Overall Passing (The percentage of students that passed math and reading.)
pass_reading_math=school_data_complete.loc[(school_data_complete["reading_score"]>=70)&(school_data_complete["math_score"]>=70)]

school_summary["Overall Passing"]=pass_reading_math.groupby(["school_name","type"]).count()["Student ID"]
school_summary["Overall Passing"]=round((school_summary["Overall Passing"]/school_summary["Total Student"])*100,4)
school_summary
```

```python
School_name=school_summary.index.get_level_values(0)
school_type=school_summary.index.get_level_values(1)
school_summary_result={"School Name":School_name,
                      "School Type":school_type,
                      "Total Students":school_summary["Total Student"],
                      "Total School Budget":school_summary["Total School Budget"],
                      "Per Student Budget":school_summary["Per Student Budget"],
                      "Average Math Score":school_summary["Average Math Score"],
                      "Average Reading Score":school_summary["Average Reading Score"],
                      "% Passing Math":school_summary["Pass Math student"],
                      "% Passing Reading":school_summary["Pass Reading student"],
                      "% Overall Passing":school_summary["Overall Passing"]
    
}
school_summary_df=pd.DataFrame(school_summary_result)
school_summary_df=school_summary_df.reset_index(drop=True)
school_summary_df=school_summary_df.set_index("School Name")
school_summary_df
```
